```python
df['gender']=df['gender'].fillna('unknown')
print(df['gender']. isnull().sum())
```

```python
df['company_type']=df['company_type'].fillna('unknown')
print(df['company_type'].isnull().sum())
```

```python
df['company_size']=df['company_size'].fillna('unknwon')
print(df['company_size'].isnull().sum())
```

```python
df['enrolled_university']=df['enrolled_university'].fillna(df['enrolled_university'].mode()[0])
print(df['enrolled_university']. isnull().sum())
```

```python
df['education_level']=df['education_level'].fillna(df['education_level'].mode()[0])
print(df['education_level'].isnull().sum())
```

```python
df['major_discipline']=df['major_discipline'].fillna(df['major_discipline'].mode()[0])
print(df['major_discipline'].isnull().sum())
```

```python
df['experience']=df['experience'].fillna(df['experience'].mode()[0])
print(df['experience'].isnull().sum())
```

```python
df['last_new_job']=df['last_new_job'].fillna(df['last_new_job'].mode()[0])
print(df['last_new_job'].isnull().sum())
```

```python
df.duplicated().sum()
```

```python
df.drop('enrollee_id',axis=1,inplace=True )
df.head()
```

```python
df['gender'] = df['gender'].fillna('unknown')


df['gender'] = df['gender'].map({'Male': 1, 'Female': 0, 'unknown': 2})


df.head()
```

```python
df['relevent_experience'] = df['relevent_experience'].fillna('Has relevent experience')
df['relevent_experience'] = df['relevent_experience'].map({'Has relevent experience': 1, 'No relevent experience': 0})
df.head()
```

```python
df['education_level'] = df['education_level'].fillna('Graduate')
df['education_level'] = df['education_level'].map({'Primary School': 0, 'High School': 1, 'Graduate': 2, 'Masters': 3, 'Phd': 4})
df.head()
```

```python
df['enrolled_university'] = df['enrolled_university'].fillna('no_enrollment')
df['enrolled_university'] = df['enrolled_university'].map({'no_enrollment': 0, 'Part time course': 1, 'Full time course': 2})
df.head()
```

```python
df.head()
```

```python
df['experience'] = df['experience'].replace({'>20': '21', '<1': '0'})
df['experience'] = df['experience'].astype(int)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['major_discipline'] = le.fit_transform(df['major_discipline'])
df.head()
```

```python
df['company_size'] = le.fit_transform(df['company_size'])
df['company_type'] = le.fit_transform(df['company_type'])
df.head()
```

```python
df['last_new_job'] = le.fit_transform(df['last_new_job'])
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['city'] = le.fit_transform(df['city'])
df.head()
```

```python
import pandas as pd
df= pd.read_csv('/content/aug_train.csv')
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
plt.figure(figsize=(5,5))

df['gender'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['#069AF3', '#FF81C0', '#929591'],
    pctdistance=0.8,
    explode=[0, 0, 0.1]
)

plt.title('Gender Distribution')
plt.ylabel('')
counts = df['gender'].value_counts()

plt.legend(
    counts.index,
    title="Segments",
    loc="upper right",
    bbox_to_anchor=(1.2, 1)
)
plt.show()
```

```python
plt.figure(figsize=(5,5))

df['relevent_experience'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('Set2')
)

plt.title('Relevant Experience Distribution')
plt.ylabel('')

counts=df['relevent_experience'].value_counts()
plt.legend(counts.index, title="Segments", loc="center left", bbox_to_anchor=(1.2, 1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,6))

df['enrolled_university'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors= ['#DDA0DD', '#580F41', '#800080'] # غيرنا الألوان بس
)

plt.title('Enrolled University Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(7, 7))

counts = df['company_type'].value_counts()

counts.plot(
    kind='pie',
    autopct='%1.0f%%',
    startangle=140,
    colors=sns.color_palette('pastel'),
    labels=None,
    pctdistance=0.8
)

plt.title('Company Type Distribution', fontsize=15, weight='bold')
plt.ylabel('')

plt.legend(
    counts.index,
    title="Segments",
    loc="upper right",
    bbox_to_anchor=(1.3, 1)
)

plt.show()

```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='education_level', hue='target', palette='Set2')
plt.title('Education Level vs Target')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

palette = ['#069AF3', '#929591']

sns.countplot(
    data=df,
    x='major_discipline',
    hue='target',
    palette=palette
)

plt.title('Major Discipline vs Target')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))

# ترتيب القيم
order = sorted(df['experience'].dropna().unique())

palette = ['#1E88E5', '#E53935']

sns.countplot(
    data=df,
    x='experience',
    hue='target',
    order=order,
    palette=palette
)

plt.title('Experience Years vs Target')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
palette = ['#4CAF50', '#929591']
sns.countplot(data=df, x='company_size', hue='target', palette=palette)
plt.title('Company Size vs Target')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
palette = ['#000000', '#E53935']
sns.countplot(data=df, x='last_new_job', hue='target', palette=palette)
plt.title('Last Job Change vs Target')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.histplot(df['training_hours'], kde=True)
plt.title('Histogram of Training Hours')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x='target', y='training_hours')
plt.title('Violin Plot: Training Hours vs Target')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='target', y='experience')
plt.title('Boxplot: Experience vs Target')
plt.show()
```

```python
sns.pairplot(
    df[['training_hours', 'experience', 'city_development_index', 'target']],
    hue='target'
)
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
